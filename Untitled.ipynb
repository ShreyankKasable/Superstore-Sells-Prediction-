```python
# as dtype of Item_Weight is float we can fill the null value with mean

df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
# as dtype of Outlet_Sizeis object so we can fill the null value with mode
# The data type of outlier size parameter is object so we will going to feel the null values with mode of that particular parameter 
# for that we are going to import askel on computer library and using that we will fill the null values

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df.iloc[:,8:9].values)
x = imputer.transform(df.iloc[:,8:9].values)
df.iloc[:,8:9] = x
```

In about two plots, we have tried to find out some correlation between outlier location type and 
item type parameter But from this, we get to know that there is no correlation between these two parameters every particular
store sells an equal number of items of that particular type with respect to the number of stores that come 
under that particular type or particular location type

```python
Outlet = list(df['Item_Type'].unique())

for i in Outlet:
    
    df_temp = df[df['Item_Type'] == i]
    plt.figure(figsize=(4,3))
    sns.countplot(data=df_temp, x = 'Outlet_Location_Type',order=df['Outlet_Location_Type'].value_counts().index)
    
    plt.xticks(rotation = 90)
    
    plt.xlabel('Item Type')
    plt.ylabel('Count')
    plt.title(f'Item type count w.r.t {i}')
    
    ax = plt.gca()
    
    for i in ax.patches:
        
        ax.text(i.get_x() + i.get_width() / 2., i.get_height(), '%d' % int(i.get_height()), fontsize = 10, ha='center', va ='bottom')
    
    plt.show()
```

Thus, in Identity item type and outlier location type there is no correlation
